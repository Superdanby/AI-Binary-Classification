```python
# Correlations to "click" are all very low.
# We should kick out some trash categories which has almost nothing to do with "click".
# Note the special case with "dclkVerticals", which has lots NaN entry. There are 2 ways to cope with it. Remove it or make the values 0 for NaN, otherwise 1.
# Categories that are probably no use( abs() < 0.01 ): deviceType, spaceType, dclkVerticals, adx, ip, os(?)

df_drop = all_df.drop(['deviceType', 'spaceType', 'dclkVerticals', 'adx', 'ip'], axis = 1).loc[:,:]
res_df = res_df.drop(['deviceType', 'spaceType', 'dclkVerticals', 'adx', 'ip'], axis = 1).loc[:,:]
res_df['click']=-1
print(res_df.size)
print(df_drop.size)
df_drop = df_drop.append(res_df)
```

```python
le = LabelEncoder()
df_le = df_drop.iloc[:,:]
for x in df_le.columns.values[:-1]:
#     print(df_labeled[x].head(10))
#     le.fit(df_labeled[x])
    df_le[x] = le.fit_transform(df_le[x].astype(str))
df_labeled = df_le[df_le['click'] != -1].loc[:,:]
df_Ylabeled = df_labeled['click'].loc[:]
df_labeled = df_labeled.drop(['click'], axis = 1).loc[:,:]
df_pred = df_drop[df_drop['click'] == -1].loc[:,:]
X_train, X_test, Y_train, Y_test = train_test_split(df_labeled, df_Ylabeled, test_size = 0.2, stratify = df_Ylabeled)
```

```python
unq, cnt = np.unique(df_Ylabeled, return_counts=True)
dict(zip(unq, cnt))
```

```python
clf = RandomForestClassifier(n_estimators=100, n_jobs=-2, class_weight={0:1, 1:10})
clf.fit(X_train, Y_train)
```

```python
tn, fp, fn, tp = confusion_matrix(clf.predict(X_test), Y_test).ravel()
(tn, fp, fn, tp)
```

```python
print((tp + tn) / (tp + tn + fp + fn))
print(classification_report(clf.predict(X_test), Y_test))
```

```python
importances = clf.feature_importances_
indices = np.argsort(importances)[::-1]
s = 0
for f in range(X_test.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 60, df_labeled.columns[:][f], importances[indices[f]]))
    s = importances[indices[f]] + s
print(s)
```

```python
df_final = df_drop.iloc[:,:]
df_final = pd.get_dummies(df_final, columns=list(df_final)[:-1])
```

```python
df_ohe = df_final[df_final['click'] != -1].loc[:,:]
df_Yohe = df_ohe['click'].loc[:]
df_Xohe = df_ohe.drop(['click'], axis = 1).loc[:,:]
df_preohe = df_final[df_final['click'] == -1].loc[:,:]
df_preohe = df_preohe.drop(['click'], axis = 1).loc[:,:]
X_trohe, X_tsohe, Y_trohe, Y_tsohe = train_test_split(df_Xohe, df_Yohe, test_size = 0.2, stratify = df_Yohe)
sc = StandardScaler()
sc.fit(X_trohe)
X_trstd = sc.transform(X_trohe)
X_tsstd = sc.transform(X_tsohe)
X_resstd = sc.transform(df_preohe)
```

```python
# LDA
lda = LinearDiscriminantAnalysis(n_components=100)
X_trlda = lda.fit_transform(X_trstd, Y_train)
X_tslda = lda.transform(X_tsstd)
X_reslda = lda.transform(X_resstd)
lda.explained_variance_ratio_
```

```python
ldarf = RandomForestClassifier(n_estimators=100, n_jobs=-2, class_weight={0:1, 1:10})
ldarf.fit(X_trlda, Y_train)
```

```python
tn, fp, fn, tp = confusion_matrix(ldarf.predict(X_tslda), Y_test).ravel()
(tn, fp, fn, tp)
```

```python
print((tp + tn) / (tp + tn + fp + fn))
print(classification_report(ldarf.predict(X_tslda), Y_test))
```

```python
pos = df_final[df_final['click'] == 1].sample(frac=1)
pos = np.split(pos, [df_final['click'].value_counts()[1] // 10 * 2], axis = 0)
train = pos[1]
# train = train.append([pos[1]] * 99, ignore_index=True)
test = pos[0]
neg = df_final[df_final['click'] == 0].sample(frac=1)
neg = np.split(neg, [df_final['click'].value_counts()[0] // 100 * 2], axis = 0)
train = train.append(neg[1], ignore_index=True).sample(frac=1)
test = test.append(neg[0], ignore_index=True).sample(frac=1)
X_train = train.drop(['click'], axis = 1)
Y_train = train['click']
X_test = test.drop(['click'], axis = 1)
Y_test = test['click']
sc = StandardScaler()
sc.fit(X_train)
X_trstd = sc.transform(X_train)
X_tsstd = sc.transform(X_test)
pca = PCA(n_components = 10)
X_trpca = pca.fit_transform(X_trstd)
X_tspca = pca.transform(X_tsstd)
pca.explained_variance_ratio_
```

```python
param_grid = [
#     {'C': [0.01, 0.1, 1, 10, 100], 'kernel': ['linear']},
   {'C': [0.01, 0.1, 1, 10, 100], 'gamma': [0.001, 0.01, (1 / 6)], 'kernel': ['rbf'], 'class_weight': [{0:1, 1:10}]},
#    {'C': [0.01, 0.1, 1, 10, 100], 'kernel': ['poly']},
#    {'C': [0.01, 0.1, 1, 10, 100], 'kernel': ['sigmoid']},
 ]
clf = GridSearchCV(SVC(), param_grid, n_jobs = -2, scoring = 'f1')
clf.fit(X_trpca, Y_train)
```

```python
tn, fp, fn, tp = confusion_matrix(clf.predict(X_tspca), Y_test).ravel()
print((tp + tn) / (tp + tn + fp + fn))
print(classification_report(clf.best_estimator_.predict(X_tspca), Y_test))
```
